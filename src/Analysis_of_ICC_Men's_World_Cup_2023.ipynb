```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### **BATTING**

```python
df = pd.read_csv("/content/drive/MyDrive/Research Materials/Datasets/ICC Men's World Cup 2023/Batting_Stats.csv")
```

```python
df
```

```python
df.columns
```

```python
runs_75th_percentile = np.percentile(df['Runs'], 75)

print("75th Percentile of 'Runs':", runs_75th_percentile)
```

```python
# Filter the DataFrame to select players who scored more runs than the 75th percentile
high_run_scorers = df[df['Runs'] > runs_75th_percentile]

# Display the players who scored more runs than the 75th percentile
print("Players who scored more runs than the 75th percentile:")
print(high_run_scorers[['Player', 'Runs']])
```

```python
print(high_run_scorers[['Player', 'Country', 'Runs', 'SR']])
```

```python
# Sort the DataFrame by 'SR' in descending order
high_run_scorers_sorted = high_run_scorers.sort_values(by='SR', ascending=False)

# Display the sorted DataFrame
print("Players who scored more runs than the 75th percentile, sorted by 'SR':")
print(high_run_scorers_sorted[['Player', 'Country', 'Runs', 'SR']])
```

```python
# Sort the DataFrame by 'Country' and 'SR'
high_run_scorers_sorted = high_run_scorers.sort_values(by=['Country', 'SR'], ascending=[True, False])

# Get unique countries and assign a color to each country
unique_countries = high_run_scorers_sorted['Country'].unique()
country_colors = sns.color_palette("husl", n_colors=len(unique_countries), as_cmap=False)

# Create a dictionary to map countries to colors
country_color_mapping = dict(zip(unique_countries, country_colors))

# Map the colors to the 'Country' column
high_run_scorers_sorted['Color'] = high_run_scorers_sorted['Country'].map(country_color_mapping)

# Plot the data with colored bars
plt.figure(figsize=(12, 6))
bars = plt.bar(high_run_scorers_sorted['Player'], high_run_scorers_sorted['SR'], color=high_run_scorers_sorted['Color'])

# Add legend with country colors
legend_labels = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=country)
                 for country, color in country_color_mapping.items()]
plt.legend(handles=legend_labels, title='Country')

plt.xlabel('Player')
plt.ylabel('Strike Rate (SR)')
plt.title('Players with Runs > 75th Percentile, Sorted by Country and SR, Colored by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by 'Country' and sum the number of sixes for each country
sixes_by_country = df.groupby('Country')['6s'].sum().reset_index()

sixes_by_country_sorted = sixes_by_country.sort_values(by='6s', ascending=False)

# Display the sorted result
print("Number of Sixes by Country (Sorted from Highest to Lowest):")
print(sixes_by_country_sorted)
```

```python
# Set seaborn color palette with different colors for each country
sns.set_palette("husl")

# Plot the data with a bar chart
plt.figure(figsize=(12, 6))
bars = sns.barplot(x='Country', y='6s', data=sixes_by_country_sorted)
plt.xlabel('Country')
plt.ylabel('Number of Sixes')
plt.title('Number of Sixes by Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plot the data with a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sixes_by_country_sorted['6s'], labels=sixes_by_country_sorted['Country'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribution of Sixes by Country')
plt.show()
```

```python
# Group by 'Country' and sum the number of fours for each country
fours_by_country = df.groupby('Country')['4s'].sum().reset_index()

fours_by_country_sorted = fours_by_country.sort_values(by='4s', ascending=False)

# Display the sorted result
print("Number of Fours by Country (Sorted from Highest to Lowest):")
print(fours_by_country_sorted)
```

```python
# Set seaborn color palette with different colors for each country
sns.set_palette("husl")

# Plot the data with a bar chart
plt.figure(figsize=(12, 6))
bars = sns.barplot(x='Country', y='4s', data=fours_by_country_sorted)
plt.xlabel('Country')
plt.ylabel('Number of Fours')
plt.title('Number of Fours by Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plot the data with a pie chart
plt.figure(figsize=(8, 8))
plt.pie(fours_by_country_sorted['4s'], labels=fours_by_country_sorted['Country'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribution of fours by Country')
plt.show()
```

```python
Highest_Wicket_Takers = df.sort_values(by='Wkts', ascending=False)[['Bowlers', 'Country', 'Wkts', 'Average', 'Economy', 'Strike Rate']]
# Display the top 5 rows in a colorful table with a title
table_html = tabulate(Highest_Wicket_Takers.head(5), headers='keys', tablefmt='html', showindex=False)
styled_table = f'<h2>Most Wickets</h2>{table_html}'
display(HTML(styled_table))
```

```python
# Filter the DataFrame to include only those bowlers who have bowled more than 19 overs
filtered_df = df[df['Overs'] > 19]

# Sort the filtered DataFrame by 'Economy' in ascending order and select the top 5 rows
top5_lowest_economy_bowlers = filtered_df.sort_values(by='Economy').head(5)


# Display the top 5 bowlers with the lowest economy in a tabular format
table_html = tabulate(top5_lowest_economy_bowlers[['Bowlers', 'Country', 'Wkts', 'Economy', 'Average', 'Strike Rate']],
                      headers='keys', tablefmt='html', showindex=False)
styled_table = f'<h2>Lowest Economy</h2>{table_html}'
display(HTML(styled_table))
```

```python
from IPython.display import display, HTML
from tabulate import tabulate

# Filter the DataFrame
filtered_df = df[df['Overs'] > 19][['Bowlers', 'Country', 'Wkts', 'Average', 'Economy', 'Strike Rate']]

# Sort the DataFrame by 'Average' in ascending order
sorted_df_lowest_average = filtered_df.sort_values(by='Average', ascending=True)

# Display the top 5 rows in a colorful table with a title
table_html = tabulate(sorted_df_lowest_average.head(5), headers='keys', tablefmt='html', showindex=False)
styled_table = f'<h2>Best Average</h2>{table_html}'
display(HTML(styled_table))
```

```python
from IPython.display import display, HTML
from tabulate import tabulate

# Filter and sort the DataFrame
filtered_sorted_df = df[df['Overs'] > 19][['Bowlers', 'Country', 'Wkts', 'Economy', 'Average', 'Strike Rate']].sort_values(by='Strike Rate', ascending=True)

# Display the table with a title
table_html = tabulate(filtered_sorted_df.head(5), headers='keys', tablefmt='html', showindex=False)
styled_table = f'<h2>Lowest Strike Rate</h2>{table_html}'
display(HTML(styled_table))
```

```python


# Group by 'Country' and sum the 'Wkts' for each country
total_wickets_by_country = df.groupby('Country')['Wkts'].sum()

# Convert the Series to a DataFrame
total_wickets_df = total_wickets_by_country.reset_index(name='Total Wickets')

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(total_wickets_df['Country'], total_wickets_df['Total Wickets'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Wickets')
plt.title('Total Wickets by Country')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Calculate the average economy for each country
average_economy_by_country = df.groupby('Country')['Economy'].mean()

# Convert the Series to a DataFrame
average_economy_df = average_economy_by_country.reset_index(name='Average Economy')

# Plotting the bar chart with y-values on top of each bar
plt.figure(figsize=(10, 6))
bars = plt.bar(average_economy_df['Country'], average_economy_df['Average Economy'], color='lightgreen')

# Annotate each bar with its y-value
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.xlabel('Country')
plt.ylabel('Average Economy')
plt.title('Average Economy by Country')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Extract the numeric part before the '/' in the 'Score' column and convert it to numeric
df['Numeric_Score'] = df['Score'].str.split('/').str[0].astype(float)

# Sort the DataFrame by 'Numeric_Score' in descending order and select the top 5 rows
top5_highest_scores = df.sort_values(by='Numeric_Score', ascending=False).head(5)

# Display the table with a title
table_html = tabulate(top5_highest_scores.head(5), headers='keys', tablefmt='html', showindex=False)
styled_table = f'<h2>Highest Scores</h2>{table_html}'
display(HTML(styled_table))

```

```python
import matplotlib.pyplot as plt

# Extract the numeric part before the '/' in the 'Score' column and convert it to numeric
df['Numeric_Score'] = df['Score'].str.split('/').str[0].astype(float)

# Group by 'Team' and sum the 'Numeric_Score' for each team
total_score_by_country = df.groupby('Team')['Numeric_Score'].sum()

# Convert the Series to a DataFrame
total_score_df = total_score_by_country.reset_index(name='Total Runs')

# Plotting the bar chart with y-values on top of each bar
plt.figure(figsize=(10, 6))
bars = plt.bar(total_score_df['Team'], total_score_df['Total Runs'], color='skyblue')

# Annotate each bar with its y-value
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.xlabel('Country')
plt.ylabel('Total Runs')
plt.title('Total Runs by Country')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert 'Run Rate' column to numeric
df['Run Rate'] = pd.to_numeric(df['Run Rate'], errors='coerce')

# Group by 'Team' and calculate the mean of 'Run Rate' for each team
average_run_rate_by_country = df.groupby('Team')['Run Rate'].mean()

# Convert the Series to a DataFrame
average_run_rate_df = average_run_rate_by_country.reset_index(name='Average Run Rate')

# Plotting the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(average_run_rate_df['Team'], average_run_rate_df['Average Run Rate'], color='lightcoral')

# Annotate each bar with its y-value
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.xlabel('Country')
plt.ylabel('Average Run Rate')
plt.title('Average Run Rate by Country')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Display the plot
plt.show()
```
