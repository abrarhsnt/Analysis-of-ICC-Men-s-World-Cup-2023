```python
import pandas as pd

# Path to your dataset folder and file
file_path = '/content/drive/MyDrive/Research Materials/Researches/Analysis of ICC Mens World Cup 2023/Dataset/ICC Mens World Cup 2023/Batting_Stats.csv'

# Load the CSV into a DataFrame
df2 = pd.read_csv(file_path)


print(df2.info())
print(df2.head())
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assuming df2 is already loaded with Batting_Stats.csv
# Select relevant columns for ranking
columns_to_consider = ['Runs', 'Average', 'SR', '100', '50', '4s', '6s']

# Handle non-numeric columns (e.g., 'High_Score', 'Player', 'Country')
df2_clean = df2[['Player', 'Country'] + columns_to_consider]

# Convert 'Average' to numeric by removing non-numeric characters (if any) and handling errors
df2_clean['Average'] = pd.to_numeric(df2_clean['Average'], errors='coerce')

# Normalize the data (Min-Max scaling)
scaler = MinMaxScaler()
normalized_values = scaler.fit_transform(df2_clean[columns_to_consider])

# Create a new DataFrame with normalized values
df_normalized = pd.DataFrame(normalized_values, columns=columns_to_consider)

# Sum the normalized values to get a composite score
df_cleaned_with_scores = df2_clean.copy()
df_cleaned_with_scores['Score'] = df_normalized.sum(axis=1)

# Rank players based on the composite score
df_cleaned_with_scores['Rank'] = df_cleaned_with_scores['Score'].rank(ascending=False, method='min')

# Sort players by rank
top_10_players = df_cleaned_with_scores.sort_values(by='Rank').head(10)

# Display the top 10 players with their rank and scores
print("Top 10 Players based on their performance:")
print(top_10_players[['Rank', 'Player', 'Country', 'Runs', 'Average', 'SR', '100', '50', '4s', '6s', 'Score']])


```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the batting and bowling datasets (df1 and df2 should already be loaded)
batting_data = df2.copy()
bowling_data = df1.copy()

# Normalize batting metrics
batting_metrics = ['Runs', 'Average', 'SR', '100', '50']
batting_data[batting_metrics] = batting_data[batting_metrics].apply(pd.to_numeric, errors='coerce')
batting_scaler = MinMaxScaler()
batting_data_normalized = pd.DataFrame(batting_scaler.fit_transform(batting_data[batting_metrics]), columns=batting_metrics)

# Add normalized batting score
batting_data['Batting_Score'] = batting_data_normalized.sum(axis=1)

# Normalize bowling metrics
bowling_metrics = ['Wkts', 'Economy', 'Strike Rate']
bowling_data[bowling_metrics] = bowling_data[bowling_metrics].apply(pd.to_numeric, errors='coerce')
bowling_scaler = MinMaxScaler()
bowling_data_normalized = pd.DataFrame(bowling_scaler.fit_transform(bowling_data[bowling_metrics]), columns=bowling_metrics)

# Calculate bowling score (inverted Economy and Strike Rate for "better is lower")
bowling_data_normalized['Economy'] = 1 - bowling_data_normalized['Economy']
bowling_data_normalized['Strike Rate'] = 1 - bowling_data_normalized['Strike Rate']
bowling_data['Bowling_Score'] = bowling_data_normalized.sum(axis=1)

# Merge batting and bowling datasets on 'Player' and 'Country'
merged_data = pd.merge(
    batting_data[['Player', 'Country', 'Batting_Score']],
    bowling_data[['Bowlers', 'Country', 'Bowling_Score']],
    left_on=['Player', 'Country'], right_on=['Bowlers', 'Country'],
    how='inner'
)

# Calculate All-Rounder Score
merged_data['All_Rounder_Score'] = merged_data['Batting_Score'] + merged_data['Bowling_Score']

# Rank players based on All-Rounder Score
merged_data['Rank'] = merged_data['All_Rounder_Score'].rank(ascending=False, method='min')

# Get the top 10 all-rounders
top_10_all_rounders = merged_data.sort_values(by='Rank').head(10)

# Display results
print(top_10_all_rounders[['Rank', 'Player', 'Country', 'Batting_Score', 'Bowling_Score', 'All_Rounder_Score']])
```

```python
import matplotlib.pyplot as plt

# Prepare data for the table plot
top_10_all_rounders_table = top_10_all_rounders[
    ['Rank', 'Player', 'Country', 'Batting_Score', 'Bowling_Score', 'All_Rounder_Score']
]

# Plotting the data as a table
plt.figure(figsize=(12, 6))
plt.gca().axis('off')  # Turn off the axis

# Create a table plot
table = plt.table(cellText=top_10_all_rounders_table.values,
                  colLabels=top_10_all_rounders_table.columns,
                  cellLoc='center',
                  loc='center',
                  colColours=["#f5f5f5"] * len(top_10_all_rounders_table.columns))

# Adjust layout for better appearance
table.auto_set_font_size(False)
table.set_fontsize(10)
table.auto_set_column_width(col=list(range(len(top_10_all_rounders_table.columns))))

plt.tight_layout()

# Save the table as a PNG file
all_rounders_png_path = "top_10_all_rounders.png"
plt.savefig(all_rounders_png_path, dpi=300)
plt.show()

print(f"Table saved as {all_rounders_png_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure that relevant columns (only numeric) are converted to numeric for ranking
batting_data['Batting_Score'] = pd.to_numeric(batting_data['Batting_Score'], errors='coerce')
bowling_data['Bowling_Score'] = pd.to_numeric(bowling_data['Bowling_Score'], errors='coerce')

# Filter players who have bowled at least 20 overs (ensure 'Overs' is numeric)
filtered_bowling_data = bowling_data[bowling_data['Overs'] >= 20]

# Rank players based on Batting Score (ascending order: best is 1)
batting_data['Batting_Rank'] = batting_data['Batting_Score'].rank(ascending=False, method='min')

# Rank players based on Bowling Score (ascending order: best is 1) for those who bowled at least 20 overs
filtered_bowling_data['Bowling_Rank'] = filtered_bowling_data['Bowling_Score'].rank(ascending=False, method='min')

# Merge batting and filtered bowling ranks, ensuring we merge on the correct columns
merged_data = pd.merge(
    batting_data[['Player', 'Country', 'Batting_Rank']],
    filtered_bowling_data[['Bowlers', 'Country', 'Bowling_Rank']],
    left_on=['Player', 'Country'], right_on=['Bowlers', 'Country'],
    how='inner'
)

# Add the sum of Batting and Bowling Ranks
merged_data['Rank_Sum'] = merged_data['Batting_Rank'] + merged_data['Bowling_Rank']

# Sort players by Rank Sum in ascending order
all_rounder_ranked = merged_data.sort_values(by='Rank_Sum').head(10)

# Prepare the table for displaying
top_10_all_rounders = all_rounder_ranked[['Player', 'Country', 'Batting_Rank', 'Bowling_Rank', 'Rank_Sum']]

# Plot the table using seaborn heatmap (no need to save, just show)
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Select only numeric columns for the heatmap
numeric_columns = ['Batting_Rank', 'Bowling_Rank', 'Rank_Sum']
heatmap_data = top_10_all_rounders.set_index('Player')[numeric_columns].T # Exclude 'Country' column

ax = sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='coolwarm', cbar=False)


# Set the title and labels
plt.title('Top 10 All-Rounders Based on Batting and Bowling Rank Sum (Bowled > 20 Overs)', fontsize=16)
plt.xlabel('Players')
plt.ylabel('Attributes')

# Show the plot in Colab
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset (replace with your actual file path)
df = pd.read_csv('/content/drive/MyDrive/Research Materials/Researches/[Current] Analysis of ICC Men’s World Cup 2023/Dataset/CSV/ICC World Cup 2023.csv')

# Define a function to apply a color scale to the table
def colorize_table(val):
    if isinstance(val, (int, float)):
        # Colorize the values on a scale from white (low) to green (high)
        color = plt.cm.viridis(val / max(df.max()))  # Normalize and apply colormap
        return f'background-color: rgba({int(color[0]*255)}, {int(color[1]*255)}, {int(color[2]*255)}, 0.7)'  # RGBA color
    else:
        return ''  # No color for non-numeric values like strings

# Apply the colorize function to the entire dataframe
styled_df = df.style.applymap(colorize_table)

# Plotting the table
fig, ax = plt.subplots(figsize=(12, 8))  # Adjust the figure size as needed
ax.axis('tight')
ax.axis('off')

# Instead of using styled_df.render(), use styled_df.to_html() to get the HTML representation
# and then extract the background colors from it.
html = styled_df.to_html()
cell_colors = [color.split(';')[0].split(':')[1].strip() for color in html.split('background-color:') if color.startswith('#')]

# Reshape cell_colors to match the shape of the table
cell_colors = np.array(cell_colors).reshape(df.shape)

# Draw the table with a style
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center', colColours=["#4CAF50"]*len(df.columns), cellColours=cell_colors)
plt.savefig('icc_world_cup_2023_colored_table_v2.png', bbox_inches='tight', dpi=300)

# Show the table
plt.show()
```
